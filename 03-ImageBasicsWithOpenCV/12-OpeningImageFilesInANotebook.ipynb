```python
img = cv2.imread('../DATA/00-puppy.jpg')
```

```python
type(img)
img.shape
```

```python
plt.imshow(img)
```

The image looks weird because openCv expects colors to be BGR instead of RGV expected by matplot.
We need to transform it.

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

## Reading grayscale images
With openCV we can read directly greyscale images. A gray scale image does not have three components.

```python
img_gray = cv2.imread('../DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap='gray')
```

## Resize images

```python
fix_img.shape
```

```python
new_img = cv2.resize(fix_img, (1000, 400))
```

```python
plt.imshow(new_img)
```

```python
w_ratio = 0.5
h_ratio = 0.5
new_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

## Flip images

```python
new_img = cv2.flip(fix_img, 0)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, -1)
plt.imshow(new_img)
```

## Saving images

```python
cv2.imwrite('totally_new.jpg', fix_img)
```

## Limiting the size of the displayed image in the notebook
